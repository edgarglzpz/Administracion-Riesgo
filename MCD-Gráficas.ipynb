```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
```

```python
df=pd.read_excel('MCD.xlsx', header=1, sheet_name='Datos')
```

```python
df.head()
```

```python
df['Rendimiento MCD'].isnull().sum()
```

```python
df.shape
```

```python
fig, (ax1, ax2) = plt.subplots(2,1, figsize=[12,5], sharex='col', sharey='row',  gridspec_kw={'hspace': 0, 'wspace': 0})
ax2.plot(df['Precio cierre ajustado S&P'], alpha=2, color='blue')
ax2.set(ylabel='S&P 500')
ax2.set_xticks(np.arange(0,650,50))
ax1.plot(df['Precio cierre ajustado MCD'], alpha=2, color='blue')
ax1.set(ylabel='MCD')
plt.savefig('Precios.png')
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(2,1, figsize=[12,5], sharex='col', sharey='row',  gridspec_kw={'hspace': 0, 'wspace': 0})
ax2.plot(df['Rendimiento S&P'], alpha=2, color='blue')
ax2.set_xticks(np.arange(0,650,50))
ax2.set(ylabel='S&P 500')
ax1.plot(df['Rendimiento MCD'], alpha=2, color='blue')
ax1.set(ylabel='MCD')
plt.savefig('Rendimientos.png')
plt.show()
```

```python
fig, ax2= plt.subplots(figsize=[12,5])

ax2.plot(df['Rendimiento S&P'], alpha=2, color='red', label='S&P 500', linestyle='dashed')
ax2.set(ylabel='Rendimientos')
ax2.plot(df['Rendimiento MCD'], alpha=2, color='blue', label='MCD')
ax2.set_xticks(np.arange(0,650,50))
ax2.legend(loc='upper left', fancybox=True, framealpha=1, shadow=True, borderpad=1)
plt.savefig('Comparativa.png')
plt.show()
```

```python
fig = plt.figure(figsize=[12,7])
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)

ax1.hist(df['Precio cierre ajustado MCD'], facecolor = 'b', alpha = 0.7)
ax1.set(title='AdjClose MCD', ylabel='Frecuencia')

ax2.hist(df['Precio cierre ajustado S&P'], facecolor = 'b', alpha = 0.7)
ax2.set(title = 'AdjClose S&P 500', ylabel='Frecuencia')

ax3.hist(df['Rendimiento MCD'], facecolor = 'b', alpha = 0.7)
ax3.set(title = 'Rend MCD', ylabel='Frecuencia')

ax4.hist(df['Rendimiento S&P'], facecolor = 'b', alpha = 0.7)
ax4.set(title = 'Rend S&P 500', ylabel='Frecuencia')

plt.savefig('Histogramas.png')
plt.show()
```
